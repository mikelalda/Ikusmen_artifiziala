### Procesado de datos con Pandas y Numpy

1. Carga en el entorno el «poblacionMunicipios.csv» proporcionado por el profesor respetando las tildes y caracteres españoles.

```python
import pandas as pd
import numpy as np
poblacionMunicipios = pd.read_csv('04-poblacionMunicipios.csv')
#display(poblacionMunicipios)
display(poblacionMunicipios.head())
```

2. Elimina las filas correspondientes a municipios sin población.

```python
poblacionMunicipios = poblacionMunicipios[poblacionMunicipios["Poblacion"] > 0] 
display(poblacionMunicipios.head())
display(poblacionMunicipios)
```

3. Imprime el total de la población de todos los municipios.

```python
#el método sum suma todas las tablas del dataframe
totalPopulation=poblacionMunicipios["Poblacion"].sum()
print("Población total ",totalPopulation)
```

4. Crea un DataFrame de Pandas donde la primera columna sean todas las provincias que aparecen en el archivo, la segunda columna el número total de habitantes por provincia, la tercera la desviación típica en el número total de habitantes por provincia y la cuarta el número total de municipios por provincia.

```python
listaProvinciasGby=poblacionMunicipios.groupby('Provincia')
listaProvincias=listaProvinciasGby.sum()
listaDeStd = listaProvinciasGby["Poblacion"].std()
numMunicipios=listaProvinciasGby["CodMunicipio"].count()

provincias = pd.DataFrame({'Poblacion':listaProvincias['Poblacion'],
                          'std':listaDeStd,
                          'Municipios':numMunicipios})
display(provincias.head())
provincias=provincias.reset_index();
display(provincias.head())
```

5. Carga el archivo «CP_Municipios.csv».

```python
municipios = pd.read_csv('04-CP_Municipios.csv')
display(municipios.head())
```

6. Empleando el código de provincia y el código de municipio, cruza los dos datasets comentados hasta el momento. El resultado final debe ser un DataFrame

```python
poblacionMerge = pd.merge(poblacionMunicipios,municipios,how='outer',on=('CodProvincia','CodMunicipio'),suffixes=('-poblacion','-municipios'),indicator=True)
display(poblacionMerge.head())
```

7. Genera un CSV llamado «faltan.csv» con todos los códigos postales que estén en «CP_Municipios.csv» pero que no tengan en «poblacionMunicipios.csv» ningún municipio con el que poder cruzar. Si todo cruza correctamente el archivo «faltan.csv» estará vacío. Se puede decir que «faltan.csv» son los códigos postales que se han perdido en el cruce a la hora de considerar la población

```python
print("Municipios sin CP en el fichero CP_Municipios.csv")
display(poblacionMerge[poblacionMerge["_merge"] == 'left_only'])
print("municipios con codigo postal no agregados al fichero poblacionMunicipios.csv")
faltanEnPoblacioncsv = poblacionMerge[poblacionMerge["_merge"] == 'right_only']
display(faltanEnPoblacioncsv)
toSave=faltanEnPoblacioncsv[['CodProvincia','CodMunicipio','CP','Municipio-municipios']]
print("Municipios que vamos a escribir en faltan.csv")
display(toSave)
toSave.to_csv("04-faltan.csv",index=False)
#all_merged_data = pd.merge(cp_municipios_data, poblacion_municipios_data, on=['CodProvincia', 'CodMunicipio'], how='left')
# tomamos aquellas filas que tengan NaN en la columna 'Poblacion'
#missing_cp_data = all_merged_data[np.isnan(all_merged_data['Poblacion'])]['CP']
```

8. Agrupa la información por código postal. Se quiere un DataFrame con las siguientes columnas:
    * Código postal.
    * Número de municipios que tienen dicho código postal asignado.
    * Población: se calculará como la suma de la población de todos los municipios que incluyen a dicho código postal.
    * Provincia: a la que está asignada el código postal.

```python
mergeBoth = poblacionMerge[poblacionMerge["_merge"] == 'both']
listaCPGb = mergeBoth.groupby('CP')

listaCP=listaCPGb['CodMunicipio'].count()
CP = pd.DataFrame({'NumMunicipios':listaCP,
                  'Poblacion':listaCPGb['Poblacion'].sum(),
                  'Provincia':listaCPGb['Provincia'].unique()})
CP = CP.reset_index()
display(CP.head())
display(CP[CP["NumMunicipios"] > 1])
```

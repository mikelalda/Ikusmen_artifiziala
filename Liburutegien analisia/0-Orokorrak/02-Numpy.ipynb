Podemos generar matrices con alguna ecuación matemática mas sofisticada. Por ejemplo en estos ejemplos, vamos a crear una matriz en la que cada fila sea 10 veces mayor que la anterior en el primer caso y 3 veces en el segundo caso

```python
c = np.array([[10*j+i for i in range(3)] for j in range(4)])
print(c)
c = np.array([[3*j+i for i in range(3)] for j in range(4)])
print(c)
```

```python
d = np.linspace(0,12,5) #genera 5 puntos entre 0 y 12 equidistantes
print(d)

#Recordad : es el operador de partición de una matriz. Se parte la matriz
#completa (:) y se añade una nueva dimensión.
#newaxis increase the dimension of the existing array by one more dimension
print(d.shape)
dp = d[np.newaxis,:]
print(dp)
print('Shape ',dp.shape)
dp = d[:,np.newaxis]
print(dp)
print('Shape ',dp.shape)
```

```python
#matrices dispersas con scipy
from scipy import sparse

X = np.random.random((5,6))
print(X)

V = X[X< 0.85]
print("filtrando")
print(V)

X[X < 0.85] = 0 # selecciona los elementos menores de 0.85 y asigna 0
print(X)

x_csr = sparse.csr_matrix(X) #CSR (Compress sparse row)

print(x_csr)

print(x_csr.toarray())
```

**Broadcasting**  NumPy distribuye el vector de menor dimensión sobre el largo para conseguir 
dimensiones compatibles. No siempre esp osible hacerlo, las dimensiones deben ser compatibles.

```python
#broadcasting
a = np.array([1.0,2.0,3.0])
b = np.array([2.0,2.0,2.0])
c = a*b
print(c)

print("--------")

a = np.array([[0.0,0.0,0.0],
              [10.0,10.0,10.0],
              [20.0,20.0,20.0],
              [30.0,30.0,30.0]])
b = np.array([1.0,2.0,3.0]) # no se debería poder sumar, ya que las dimensiones no son iguales. en este caso 
# Numpy lo que hace es aplicar b sobre a tantas veces como filas hay.
c = a + b
print(c)
```

```python
#Otras operaciones interesantes
np.random.seed(12345) # determina una semilla de números eleatorios.
r=np.random.rand(4)
print(r)
rint=r.astype(int)  #cambia el tipo
print(rint)
print(np.round(r)) # redondea ( >= 0.5)
```

```python
x=np.random.rand(2,2)
print(x) 
print(x.T) #Traspuesta
print(x*5) # producto por un escalar
print(x+3) # suma a todos los elementos de la matriz
print(x*x.T) # producto de la matriz por su traspuesta
print(np.dot(x,x.T))  # producto vectorial
```

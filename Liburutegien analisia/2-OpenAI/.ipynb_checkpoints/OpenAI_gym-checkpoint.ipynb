```python
import gym
env = gym.make('CartPole-v0')
iters = 20
episo=100
for i_episode in range(episo):
    R = 0
    observation = env.reset()
    for t in range(iters):
        env.render()
        action = env.action_space.sample()
        observation, reward, done, info = env.step(int(action))
        R += reward
       
        if done:
            print(observation[0])
            print("Episode finished after {} timesteps and {} total reward".format(t+1, R))
            break
env.close()
```

## Reinforcement learning

Kasu honetan Q gure algoritmoko memoria izango da, 500 egoera gordeko dituelarik eta 6 akzio egoera bakoitzean. 

```python
Q = np.zeros([env.observation_space.n, env.action_space.n])
```

```python
G = 0 # Rekonpentsa orokorra bakoitzean
```

```python
alpha = 0.618 # Ikasteko ratioa, bat aukeratzearren.
```

Ondoren doana izango da gure algoritmoa

```python
episode = 1000
for i in range(episode):
    done = False
    G, reward = 0, 0
    state = env.reset()
    count = 0
    while done != True:
            action = np.argmax(Q[state]) #1
            state2, reward, done, info = env.step(action) #2
            Q[state,action] += alpha * (reward + np.max(Q[state2]) - Q[state,action]) #3
            G += reward
            state = state2
            count+=1
    if i % 50 == 0:
        print('Episode {} Total Reward: {} Times: {}'.format(i,G,count))
```

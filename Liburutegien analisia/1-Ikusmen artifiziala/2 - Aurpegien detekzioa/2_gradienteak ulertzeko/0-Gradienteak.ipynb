# OpenCV gradienteak

Atal honetan irudien ertzak edota inguraketak egiteko gradienteak erabiliko ditugu.

Orain arte kolore zehatz batzuk begiratzen genituen eta horien trataera egiten genuen. Baina argi geratu da, kasu batzuetarako ez dela egokia prozesatzeko modu hori. Hori dela eta koloreen aldaketak nehurtuko dira eta horien aldaketa bortitzak bakarrik marraztuko ditugu. Hau egiteko gradienteak erabiliaz.

```python
import cv2
import numpy as np
```

```python
def lap_grad(image):
    laplace = cv2.Laplacian(image,cv2.CV_32F)
    return laplace
```

```python
def sob_grad(image):
    gx = cv2.Sobel(image, cv2.CV_32F, 1, 0, ksize=1)
    gy = cv2.Sobel(image, cv2.CV_32F, 0, 1, ksize=1)
    # Python Calculate gradient magnitude and direction ( in degrees ) 
    mag, angle = cv2.cartToPolar(gx, gy, angleInDegrees=True)
    return (mag, angle)
```

```python
def main():
    # Read image
    img = cv2.imread('Mikel.jpg')

    # Neurria aldatu diot handiegia ikusten delako
    scale_percent = 50 # ehuneko zenbatean jarriko den
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dim = (width, height)
    # resize image
    img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)

    img = np.float32(img) / 255.0

    # Calculate gradient
    #Laplace erabiliz
    laplace = lap_grad(img)
    #Sobel erabiliz
    mag, angle = sob_grad(img)

    cv2.imshow('Originala',img)
    cv2.imshow('Sobel',mag)
    cv2.imshow('Laplace',laplace)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if __name__ == "__main__":
    main()
```

Gauza berdina bideo batean egin nahi badugu, kodea zerbait aldatu beharko da baina erabiliko diren funtzioak berdinak izango dira.

```python
def main():
    cap = cv2.VideoCapture(0)
    # Read image

    while (True):
        # Irakurri irudia
        ret, img = cap.read()
        img = np.float32(img) / 255.0
        #Kalkulatu gradientea bi modutako batera
        sob=1
    
        if sob:
            mod = 'Sobel gradientea'
            grad, angle = sob_grad(img)
        else:
            mod = 'Laplace gradientea'
            grad = lap_grad(img)
        cv2.imshow('Irudi originala',img)
        cv2.imshow(mod,grad)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```

```python
if __name__ == "__main__":
    main()
```

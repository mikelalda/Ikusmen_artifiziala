# OpenCV - Irudiekin jokatzen

Opencv liburutegia sartuko dugu. Honekin batera irakurtzen diren pixel guztiak ikusteko, numpy sartuko dugu. Guztia matrize baten dagoelako. 

```python
import numpy as np
import cv2
```

## Orokorra

Argazki batean nola idatzi edo zuzenak, laukia edota poligonoak egin ikusteko, irudi beltz bat eraikiko dugu. Horretarako numpyrekin matrize bat osatuko dugu zeroz betea.

```python
black_img = np.zeros((512, 512, 3), np.uint8)
```

```python
cv2.imshow('BELTZA',black_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Irudi horretan zuzen bat jartzeko, non hasi (0,0) kasu honetan, non bukatu (512,512), bere kolorea (255,255,255) txuria eta lodiera (5)  jarriko diogu

```python
cv2.line(black_img, (0,0), (512,512), (255,255,255), 5)
cv2.imshow('BELTZA',black_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Lauki bat egiteko sistema berdina erabiliko dugu. Non hasi, non bukatu, kolorea eta lodiera aukeratuko ditugu.

```python
cv2.rectangle(black_img, (0,138), (220,400), (0,255,0), 2)
cv2.imshow('BELTZA',black_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Poligono bat egiterako garaian berriz, zein puntutatik pasatuko den esan beharra dugu. Horretarako beste matrize bat egingo dugu non puntu denak zehaztuko ditugun.

```python
puntuak = np.array([[60,55],[70,80],[140,80],[100,60]], np.int32)
puntuak = puntuak.reshape((-1,1,2))
cv2.polylines(black_img, [puntuak], True, (0,0,255))
cv2.imshow('BELTZA',black_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Zerbait idatzi nahi badugu honakoa egingo dugu.

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(black_img, 'GUK NAHI DUGUNA IDATZI', (10,500), font, 1, (255,0,0), 2, cv2.LINE_AA)
cv2.imshow('BELTZA',black_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Iruedien erabilera

Irudiak irakurtzeko ondorengo kodigoa sartuko dugu. Normalean bere ubikazioa sartu behar zaio

```python
img = cv2.imread('Argazkiak\input.jpg')
```

Irudiak idazterako garaian non sartuko dugun esan behar diogu eta eman diogun izenari deitu

```python
cv2.imwrite("Argazkiak/output_gray.jpg", img)
```

Irudiaren luzapena aldatzea nahi badugu, modu honetako zerbait beharko genuke.

```python
cv2.imwrite("Argazkiak\output.png", img, [cv2.IMWRITE_PNG_COMPRESSION])
```

BGR (Blue Green Red) sistema erabiltzen da askotan. Honek esan nahi duena da urdinaren balioa 0tik 256ra balio dezake eta beste biek ere berdin. Hiru zenbakiak matrize batean sartzen dira eta hauen konbinazioekin edozein kolore sortzea dago, beraz hiru zenbakirekin kolore bat definitzen da. Ordorengo kodean ikusten da matrize bat esaterako.

```python
print(img)
```

Irudia ikustea nahi badugu, imshow funtzioa erabiliko dugu. Errealitatean "cv2.imshow('irudia', img)" bakarrik beharko genuke, baina kasu honetan aplikazioa blokeatzen denez erakusten dugun irudia itxi egin behar da jarraitzeko. Hori dela eta tkla bat sakatzeko zain gelditzen da sistema eta behin hau gertatzean irudi guztiak ixteko esango diogu.

```python
cv2.imshow('irudia', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Koloreen espazioak garrantzitsuak dira ikusmen artifizialean, eragiketak egiterako garaian erraztasunak ematen dituzte. Honen adibide bezala konbertsio desberdinak ikusiko ditugu.

```python
# Gure argazkia grisera pasatzeko
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Gure argazkia hsv formatura pasatzeko
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('GRISA',gray_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.imshow('HSV',hsv_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Argazkia lausotuta jarri nahi badugu, irudiaren ondorengo parentesi arteko zenbakiak zenbat eta handiagoak izan orduan eta lausotuagoa agertuko da irudia.

```python
img_gaussian = cv2.GaussianBlur(img, (19, 19), 0)
cv2.imshow('Lausotua',img_gaussian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Argazkia dilatatu nahi badugu, hau da, pixel bakoitza dilatatu, dilated erabiliko da eta iterations aldagaian esango diogu zenba nahi dugun dilatatu

```python
img_dilated = cv2.dilate(img, None, iterations=1)
cv2.imshow('IRAGAZKIA',img_dilated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Argazkian agertuko diren koloreak limitatu nahi baditugu, limitazioa (threshold) erabiliko dugu. kasu honetan binarioa erabiliko dugu, errazena. 50 Zenbakia izango da kasu honetan filtroa egingo duena, bere balioaren gainetik dagoen edozein zenbaki txurira joango da, 255 zenbakia jarri diogulako, eta alderantziz berdina , baina beltzera.

```python
ret, img_threshold = cv2.threshold(gray_img, 50, 255, cv2.THRESH_BINARY)
cv2.imshow('IRAGAZKIA',img_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, img_threshold = cv2.threshold(gray_img, 50, 255, cv2.THRESH_BINARY_INV)
cv2.imshow('IRAGAZKIA',img_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Irudi gris bat hartu eta aurreko filtroarekin bezala egingo da lan. Baina adaptatu egingo da lausotutako filtroa erabiliaz.

```python
img_adapted_threshold = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 201, 10)
cv2.imshow('IRAGAZKIA',img_adapted_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Lausotua erabili beharrean posible da bataz-bestekoarekin egitea.

```python
img_adapted_threshold = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 201, 10)
cv2.imshow('IRAGAZKIA',img_adapted_threshold)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

__Elementu desberdinak inguratzeko modu desberdinak daude.__

Inguraketak detektatzeko findContours funtzioa erabiliko da. Funtzio honen barruan hainbat modu daude iragazki desberdinak egiteko.

1. Guztiak hartzeko RETR_LIST erabiliko da.
2. Kanpokoak bakarrik hartzeko RETR_EXTERNAL erabiliko da. Honekin kontorno bat egiten, horren barruan dauden txikiagokoei ez die kasurik egingo.
3. Guztiak hartu baina aukeraketa egiteko, RETR_CCOMP erabiliko da
4. Hierarki bat egiteko RETR_TREE erabiliko da. Honek bakoitza zein nibeletan dagoen esango digu eta bat bestearen barruan badago nibel txikiagoan agertuko da.

__Eman beharreko pausoak:__

1. gure irudia irakurri, RGB irudia.
2. Irudia grisera pasatuko dugu, argitasunaren etenak detektatzean oinarritzen delako.
3. Irudia binariora pasatuko dugu txuri-beltzean jarriz.
4. Elementuen inguraketa zehaztuko da aurreko iruditik.
5. inguraketa marrastuko da.

__Pausoak funtziotan banatuta honela izango dira__

1. Irudia irakurri.

```python
def read_rgb_image(image_name, show):
    rgb_image = cv2.imread(image_name)
    if show: 
        cv2.imshow("RGB Image",rgb_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows
    return rgb_image
```

2. Irudia grisera pasatu

```python
def convert_rgb_to_gray(rgb_image,show):
    gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_BGR2GRAY)
    if show: 
        cv2.imshow("Gray Image",gray_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows
    return gray_image
```

3. Txuri beltzeko irudia lortu. Kasu honetan bi aukera ematen dira, iragazki lausotua eta iragazki normala.

```python
def convert_gray_to_binary(gray_image, adaptive, show):
    if adaptive: 
        binary_image = cv2.adaptiveThreshold(gray_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 15, 10)
    else:
        _,binary_image = cv2.threshold(gray_image,195,255,cv2.THRESH_BINARY_INV)
    if show:
        cv2.imshow("Binary Image", binary_image)
        cv2.waitKey(0)
        cv2.destroyAllWindows
    return binary_image   
```

4. Txuri-beltzeko iruditik inguraketa eremuak atera.

```python
def getContours(binary_image):      
    contours, hierarchy = cv2.findContours(binary_image, 
                                              cv2.RETR_LIST, 
                                               cv2.CHAIN_APPROX_SIMPLE)
    return contours
```

5. Inguraketak egin

```python
def draw_contours(image, contours, image_name):
    index = -1 #inguraketa guztiak esan nahi du
    thickness = 2 #inguraketa marraren lodiera
    color = (255, 0, 255) #marraztuko den kolorea definitzeko
    cv2.drawContours(image, contours, index, color, thickness)
    cv2.imshow(image_name,image)
    cv2.waitKey(0)
    cv2.destroyAllWindows
```

Modu honetan ez dugu lortzen guk nahi dugun ingurunea detektatzeko ordea ez digu balio. Hori egin ahal izateko beste funtzio bat sortuko dugu. Funtzio honek egingo duena honakoa da, lehenik eta behin inguraketa guztiak banaka hartu eta gero hauek inguratuko ditu baldin eta bere azalera 100 baino handiagoa bada.

```python
def process_contours(contours, binary_image, rgb_image):
    # irudi errealaren neurri bereko irudi beltz bat sortuko dugu 
    black_image = np.zeros([binary_image.shape[0], binary_image.shape[1], 3], 'uint8')
    for c in contours:
        # Irudiko inguraketa erakutsiko dugu.
        # Lehenik eta behin i
        cv2.drawContours(rgb_image, [c], -1,(255,0,255),1)
        cv2.drawContours(black_image, [c],-1, (255,0,255),1)
        
        
        area = cv2.contourArea(c)
        perimeter = cv2.arcLength(c, True)
        ((x, y), radius) = cv2.minEnclosingCircle(c)
        if area >100 :
            M = cv2.moments(c)
            cx = int(M['m10']/M['m00'])
            cy = M['m01']/M['m00']
            cv2.circle(rgb_image,(int(cx),int(cy)), int(radius), (0,0,255), 1)
            cv2.circle(black_image,(int(cx),int(cy)), int(radius), (0,0,255), 1)
        
    cv2.imshow('black',black_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    cv2.imshow('rgb',rgb_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

Funtzionamendu orokorra kontrolatzeko funtzioa da hau

```python
def main():
    image_name = 'Argazkiak/tennisball01.jpg'
#     image_name = 'Argazkiak/tennisball02.jpg'
#     image_name = 'Argazkiak/tennisball03.jpg'
#     image_name = 'Argazkiak/tennisball04.jpg'
#     image_name = 'Argazkiak/tennisball05.jpg'
    rgb_image = read_rgb_image(image_name, True)
    gray_image= convert_rgb_to_gray(rgb_image,True)
    binary_image = convert_gray_to_binary(gray_image, True, True)
    contours = getContours(binary_image)
    draw_contours(rgb_image, contours,"RGB Contours")
    process_contours(contours, binary_image, rgb_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
if __name__ == "__main__":
    main()
```

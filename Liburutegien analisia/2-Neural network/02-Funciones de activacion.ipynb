```python
import numpy as np
import matplotlib.pyplot as plt
```

El error cuadratico medio:

$C=\frac{1}{2}\sum_{j}^{}(y_{j}-a^{l})^{2}$

$\frac{\partial C}{\partial a^{l}}= (a_{j}^{l}-y_{j})$

Escalonado:

$ f(x) = \binom{x< 0 \rightarrow  x= 0}{x\geq 0 \rightarrow  x= 1}$

$\frac{\partial f(x)}{\partial x}= 0$

```python
esc = (lambda x: x>0,
       lambda x: x==0 )
```

Unidad lineal rectificada (ReLU):

$ f(x) = \binom{x< 0 \rightarrow  x= 0}{x\geq 0 \rightarrow  x= x}$

$\frac{\partial f(x)}{\partial x} = \binom{x< 0 \rightarrow  x'= 0}{x\geq 0 \rightarrow  x'= 1}$

```python
relu = (lambda x: np.maximum(0, x),
       lambda x: x>0 )
```

Sigmoide:

$f(x)=\frac{1}{1+e^{x}}$

$\frac{\partial f(x)}{\partial x}= f(x)\times (1-f(x))$

```python
sigm = (lambda x: 1/ ( 1 + np.e **(-x)),
        lambda x: sigm[0](x) * (1 - sigm[0](x)))
```

Tangencial:

$f(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$

$\frac{\partial f(x)}{\partial x}= 1-f(x)$

```python
tanh = (lambda x: (np.e ** x - np.e ** (-x))/(np.e ** x + np.e ** (-x)),
       lambda x: 1-tanh[0](x))
```

```python
_x = np.linspace(-5, 5, 100)

fig = plt.figure()
#plt.title("Funciones de activacion con sus respectivas derivadas")

ax1 = fig.add_subplot(421) #4 filas y 2 columnas, primer
ax1.plot(_x, sigm[0](_x))
ax1.grid(True)
ax1.set_title('Sigmoide',fontweight="bold")

ax2 = fig.add_subplot(422) #4 filas y 2 columnas, segundo
ax2.plot(_x, sigm[1](_x))
ax2.grid(True)
ax2.set_title('Derivada de sigmoide', fontweight="bold")

ax3 = fig.add_subplot(423) #4 filas y 2 columnas, 
ax3.plot(_x, relu[0](_x))
ax3.grid(True)
ax3.set_title('Relu',fontweight="bold")

ax4 = fig.add_subplot(424) #4 filas y 2 columnas, 
ax4.plot(_x, relu[1](_x))
ax4.grid(True)
ax4.set_title('Derivada de relu', fontweight="bold")

ax5 = fig.add_subplot(425) #4 filas y 2 columnas, primer
ax5.plot(_x, tanh[0](_x))
ax5.grid(True)
ax5.set_title('Tangencial',fontweight="bold")

ax6 = fig.add_subplot(426) #4 filas y 2 columnas, segundo
ax6.plot(_x, tanh[1](_x))
ax6.grid(True)
ax6.set_title('Derivada de tangencial', fontweight="bold")

ax7 = fig.add_subplot(427) #4 filas y 2 columnas, primer
ax7.plot(_x, esc[0](_x))
ax7.grid(True)
ax7.set_title('Escalar',fontweight="bold")

ax8 = fig.add_subplot(428) #4 filas y 2 columnas, segundo
ax8.plot(_x, esc[1](_x))
ax8.grid(True)
ax8.set_title('Derivada de escalar', fontweight="bold")


plt.subplots_adjust(bottom=-0.4, top=2, left=-0.3, right=1.1)
plt.show(fig)
```

# Sare neuronala

Sare neuronala behean ikusten dugun irudian agertzen den modukoa da. Ditugun sarreak kontuan edukiz, hauek landuz eta eragiketak eginez, hau da, neurona batzuetatik pasatu ondoren, irteera bat edukiko genuke. Sistema banatzen de kapei, sarrerako kapa (input layers), kapa ezkutuak (hidden layer) eta ireteerako kapa (output layer) deitzen zaie.

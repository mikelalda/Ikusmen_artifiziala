# Sare neuronala

Sare neuronala behean ikusten dugun irudian agertzen den modukoa da. Ditugun sarreak kontuan edukiz, hauek landuz eta eragiketak eginez, hau da, neurona batzuetatik pasatu ondoren, irteera bat edukiko genuke. Sistema banatzen de kapei, sarrerako kapa (input layers), kapa ezkutuak (hidden layer) eta ireteerako kapa (output layer) deitzen zaie.

Ondrengo irudian ikusten den bezala sinplifikatuko genuke.

### Kapen azalpena

Lehenengo irudian ikusten den pelota bakoitzak modu honetan egingo luke lan.

<img src="https://miro.medium.com/max/700/1*YHQ5e963XrraPG4yh3R4Rw.png" width="400" />

$x \rightarrow  sarrerak$

$w \rightarrow  Sarrera bakoitzari emango zaion pisua$

$b \rightarrow  alborapena$

Beraz, neurona honetara iritsiko den informazioa honakoa litzateke, baldin eta neurona horri z deitzen badiogu:

$z = x_0\times w_0 + x_1\times w_1 + x_2\times w_2 + x_3\times w_3 + b$

Ikusten duzuen bezala, lehenik ikusitako ekuazio bat atera dugu, regresio linear bat. Regresio linear asko lotzen baditugu ordea gure emaitzak ekuazio linear bat izaten jarraituko du. Hori dela eta aktibazioa sartu behar diogu gure sistemari.

Aktibazio funtzioa emaitz lineal bat ez edukitzeko jarriko da, hau da, neurona guztietatik lortutako emaitza ez izateko linearra. Erabiltzen diren ekuazioak honakoak dira, eskalarra, unitate lirear rektifikatua, sigmoidea eta tangentziala.

### Zuzentzailea

Behin sistema montatuta, hasierako balioak posible da ongi ez egotea eta hauek aldatu beharra izatea. Lehenengo irudian ikusi dugun sare neuronal txiki horretan eskuz aldatzen joan jaitezke nahi dugun emaitza lortu arte, baina ez litzateke egokiena izango. Hori dela eta, sistema entrenatu egin behar dugu, hau da, eman dizkiogun pisuen eta alborapenaren balioak aldatzen joango gara automatikoki Backpropagation erabiliaz. 

Backpropagation erabiliz, kapaz kapa joango gara gure erroreak begiratzen eta neurona bakoitzaren errorea ikusiko dugu. Errore hori, gradienteen gainbehera, kontutan izanik gure pisuen eta alborapenen aldagaiak aldatzen joango gara gure sistema ongi entrenatuta egon arte. Ondorengo irudian agertzen da hori grafikoki.

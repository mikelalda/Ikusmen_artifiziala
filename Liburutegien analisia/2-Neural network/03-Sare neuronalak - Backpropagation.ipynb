# Backpropagation

Backpropagation erabiliaz gures sisteman ditugun neuronetako gastua kalkulatuko dugu. Horretarako gure sistemaren gradientea kalkulatuko dugu, hau da, gure sistemaren gasuta sarrerako aldagai bat aldatzen denean.

$\frac{\partial C}{\partial w}$  izando da matematikoki adierazita

Berdina egingo dugu alborapenarekin

$\frac{\partial C}{\partial b}$  izando da matematikoki adierazita

Gastu hau baloratzerako garaian emaitzetik hasiko ginateke begiratzen atzeraka. 

Kapetan oinarritzen garenez, azkenengo kapak bere aurrekoari eragiten dio eta horrek aurrekoari, beraz guztiaen lotura egin beharra dago atzekoz aurrera.
________________________________

__Lehenik eta behin, azkenengo kapa landuko dugu:__

Z gure neurona izango da:

$Z^L = w^L + b^L$

a gure aktibazioa:

$a(Z^L)$

Azkennik C izango dugu batazbesteko errore kuadratikoa:

$C(a(Z^L))$

Beraz esan bezala, $\frac{\partial C}{\partial w}$ egingo dugu eta hau zatitan banatzen badugu honako emaitza izango dugu matematikoki.

$\frac{\partial C}{\partial w} = \frac{\partial C}{\partial a} \times \frac{\partial a}{\partial z} \times \frac{\partial z}{\partial w}$ 

$\frac{\partial C}{\partial a} \rightarrow$ Erroreak izango duen eragina, batazbesteko errore kuadratikoaren deribatua

$\frac{\partial a}{\partial z} \rightarrow$ aktibazioak izango duen erabina, aktibazioaren deribatua

$\frac{\partial z}{\partial w} \rightarrow$ sarrerak funtzioan izango duen eragina
___

Batazbesteko errore kuadratikoa:

$C=\frac{1}{2}\sum_{j}^{}(y_{j}-a^{l})^{2}$

$\frac{\partial C}{\partial a^{l}}= (a_{j}^{l}-y_{j})$

Aktibazio funtzioaren deribatua, kasu honetan sigmoidearen adibidea jarriko dugu:

$a(z)=\frac{1}{1+e^{z}}$

$\frac{\partial a}{\partial z}= a\times (1-a)$

Gure sarrerak duen eragina begiratu nahi badugu, aurreko sarrerako aktibazioa begiratu beharko genuke, hau baita honi eragingo diona. Beraz, hau litzateke sisteman izango duen eragina.

$\frac{\partial z}{\partial a}= a^{L-1}$
___

Berdina egingo dugu alborapenarekin:

Beraz esan bezala, $\frac{\partial C}{\partial b}$ egingo dugu eta hau zatitan banatzen badugu honako emaitza izango dugu matematikoki.

$\frac{\partial C}{\partial b} = \frac{\partial C}{\partial a} \times \frac{\partial a}{\partial z} \times \frac{\partial z}{\partial b}$ 

$\frac{\partial C}{\partial a} \rightarrow$ Erroreak izango duen eragina, batazbesteko errore kuadratikoaren deribatua

$\frac{\partial a}{\partial z} \rightarrow$ aktibazioak izango duen erabina, aktibazioaren deribatua

$\frac{\partial z}{\partial b} \rightarrow$ sarrerak funtzioan izango duen eragina

Kasu honetan aldatzen den bakarra $\frac{\partial z}{\partial b}$ da, beste guztia berdin geratuko litzateke:

$\frac{\partial z}{\partial b} = 1$

$\partial^{L} = \frac{\partial C}{\partial z^{L}} \rightarrow $ Neuronari sarturiko errorea

$\frac{\partial C}{\partial w} = a^{L-1} \times \partial^{L}$

$\frac{\partial C}{\partial b} = 1 \times \partial^{L}$
___

__Azken aurreko kapa:__

 Azken aurreko kapa lantzearekin batera beste kapa guztiak landu ahal ditugu

$\frac{\partial C}{\partial w^{L-1}} = \frac{\partial C}{\partial a^{L}} \times \frac{\partial a^{L}}{\partial z^{L}} \times \frac{\partial z^{L}}{\partial a^{L-1}} \times \frac{\partial a^{L-1}}{\partial z^{L-1}} \times \frac{\partial z^{L-1}}{\partial w^{L-1}}$

Edo

$\frac{\partial C}{\partial w^{L-1}} = \partial^{L} \times \frac{\partial z^{L}}{\partial a^{L-1}} \times \frac{\partial a^{L-1}}{\partial z^{L-1}} \times \frac{\partial z^{L-1}}{\partial w^{L-1}}$

Non:

$\frac{\partial z^{L-1}}{\partial a^{L-1}} \rightarrow $ deribada de la funcion de activacion

$\frac{\partial z^{L-1}}{\partial w^{L-1}} \rightarrow a^{L-2}$ 

Beraz falta zaigun datu bakarra ondorengoa da, nola eragiten duen batura ponderatuak aurreko kapako irteera aldatzen denean.

$\frac{\partial z^{L}}{\partial a^{L-1}} \rightarrow w^{L}$



$\frac{\partial C}{\partial a^{L}} \times \frac{\partial a^{L}}{\partial z^{L}} \times \frac{\partial z^{L}}{\partial a^{L-1}} \times \frac{\partial a^{L-1}}{\partial z^{L-1}} \rightarrow = \partial^{L-1}$




$\frac{\partial C}{\partial w^{L-1}} = \partial^{L-1} \times a^{L-2}$

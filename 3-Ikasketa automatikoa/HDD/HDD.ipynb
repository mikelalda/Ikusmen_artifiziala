```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_Q4_2020/2020-10-01.csv')
```

```python
pd.set_option('display.max_columns', None)
display(df.head())
```

```python
df.describe()
```

```python
pd.set_option('display.max_rows', None)
Features = df.columns
display(df[Features].isna().sum().sort_values(ascending=True))
```

```python
fig = plt.figure(num=None, figsize=(100, 150), dpi=80)
corr = df.corr().abs()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    linewidths=.2,
    annot=True,
    fmt='.2f'
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(100, 150), dpi=80)

m = (corr.mask(np.eye(len(corr), dtype=bool)).abs() > 0.75).any()
raw = corr.loc[m, m]
ax = sns.heatmap(
    raw, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True,
    linewidths=.2,
    annot=True,
    fmt='.2f'
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
plt.show()
```

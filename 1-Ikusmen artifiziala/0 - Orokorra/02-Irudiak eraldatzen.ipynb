```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Iruedien erabilera

Irudiak irakurtzeko ondorengo kodigoa sartuko dugu. Normalean bere ubikazioa sartu behar zaio

```python
img = cv2.imread('Argazkiak\input.jpg')
```

Irudiak idazterako garaian non sartuko dugun esan behar diogu eta eman diogun izenari deitu

```python
cv2.imwrite("Argazkiak/output_gray.jpg", img)
```

Irudiaren luzapena aldatzea nahi badugu, modu honetako zerbait beharko genuke.

```python
cv2.imwrite("Argazkiak\output.png", img, [cv2.IMWRITE_PNG_COMPRESSION])
```

Koloreen espazioak garrantzitsuak dira ikusmen artifizialean, eragiketak egiterako garaian erraztasunak ematen dituzte. Honen adibide bezala konbertsio desberdinak ikusiko ditugu.

```python
# Gure argazkia grisera pasatzeko
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Gure argazkia hsv formatura pasatzeko
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(gray_img, cmap='gray')
plt.show()
plt.imshow(hsv_img, cmap='hsv')
plt.show()

hist,bins = np.histogram(gray_img.flatten(),256,[0,256])

plt.hist(img.flatten(),256,[0,256])
plt.show()
```

Argazkia lausotuta jarri nahi badugu, irudiaren ondorengo parentesi arteko zenbakiak zenbat eta handiagoak izan orduan eta lausotuagoa agertuko da irudia.

```python
img_gaussian = cv2.GaussianBlur(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY), (19, 19), 0)
plt.imshow(gray_img, cmap='gray')
plt.show()
plt.imshow(img_gaussian, cmap='gray')
plt.show()
```

```python
plt.imshow(gray_img, cmap='gray')
plt.show()

print('THRESH_BINARY')
ret, img_threshold_bin = cv2.threshold(gray_img,150,255,cv2.THRESH_BINARY)
plt.imshow(img_threshold_bin, cmap='gray')
plt.show()

print('THRESH_BINARY_INV')
ret, img_threshold_bin_inv = cv2.threshold(gray_img,150,255,cv2.THRESH_BINARY_INV)
plt.imshow(img_threshold_bin_inv, cmap='gray')
plt.show()

print('THRESH_TRUNC')
ret, img_threshold_trunc = cv2.threshold(gray_img,150,255,cv2.THRESH_TRUNC)
plt.imshow(img_threshold_trunc, cmap='gray')
plt.show()

print('THRESH_TOZERO')
ret, img_threshold_zero = cv2.threshold(gray_img,150,255,cv2.THRESH_TOZERO)
plt.imshow(img_threshold_zero, cmap='gray')
plt.show()

print('THRESH_TRUNC_TOZERO')
ret, img_threshold_zero2 = cv2.threshold(img_threshold_trunc,149,255,cv2.THRESH_TOZERO)
plt.imshow(img_threshold_zero2, cmap='gray')
plt.show()

print('THRESH_TOZERO_INV')
ret, img_threshold_zero_inv = cv2.threshold(gray_img,150,255,cv2.THRESH_TOZERO_INV)
plt.imshow(img_threshold_zero_inv, cmap='gray')
plt.show()
```

```python
ret, img_threshold = cv2.threshold(gray_img, 50, 255, cv2.THRESH_BINARY_INV)
plt.imshow(gray_img, cmap='gray')
plt.show()
plt.imshow(img_threshold, cmap='gray')
plt.show()
```

Irudi gris bat hartu eta aurreko filtroarekin bezala egingo da lan. Baina adaptatu egingo da lausotutako filtroa erabiliaz.

```python
img_adapted_threshold = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 201, 10)
plt.imshow(gray_img, cmap='gray')
plt.show()
plt.imshow(img_adapted_threshold, cmap='gray')
plt.show()
img_adapted_threshold = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 201, 10)
plt.imshow(img_adapted_threshold, cmap='gray')
plt.show()
```

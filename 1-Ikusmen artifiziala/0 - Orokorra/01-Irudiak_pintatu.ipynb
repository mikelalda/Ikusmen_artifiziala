```python
import cv2
import numpy as np
import time

drawing = False # Xagua zapalduta True
mode = 'k' # modua aukeratzeko, zer egin 
ix,iy = -1,-1 # Hasieratzea

# Xaguaren callback funtzioa
def draw_circle(event,x,y,flags,param):
    global ix,iy,drawing,mode

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == 'k':
                cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
            elif mode == 'b':
                cv2.circle(img,(x,y),5,(0,0,255),-1)
            elif mode == 'l':
                nx,ny = x,y
                cv2.line(img,(ix,iy),(x,y),(255,0,0),5)
                ix,iy = x,y

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        if mode == 'k':
            cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
        elif mode == 'b':
            cv2.circle(img,(x,y),5,(0,0,255),-1)
        elif mode == 'l':
            cv2.line(img,(ix,iy),(x,y),(255,0,0),5,-1)
            
        
```

```python
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('irudia')
cv2.setMouseCallback('irudia',draw_circle)

while(1):
    cv2.imshow('irudia',img)
    k = cv2.waitKey(1) & 0xFF
    if k == ord('k'):
        mode = 'k'
    elif k == ord('b'):
        mode = 'b'
    elif k == ord('+'):
        mode = 'l'
    elif k == 27:
        break

cv2.destroyAllWindows()
```
